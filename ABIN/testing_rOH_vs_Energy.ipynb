```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import seaborn as sns

custom_params = {"xtick.minor.visible": True,
                 "xtick.major.pad":5,
                 "xtick.direction":"in",
                 "xtick.top":True,
                 "xtick.labelsize":12.0,
                 "ytick.minor.visible": True,
                 "ytick.direction":"in",
                 "ytick.right":True,
                 "ytick.labelsize":12.0,
                 "font.family":"Helvetica",
                 "font.size":12.0,
                 "lines.linewidth":2,
                 "legend.frameon":False,
                 "legend.labelspacing":0,
                 "legend.borderpad":0.5,
                 'legend.fontsize': 12.0,}

sns.set_theme(style='ticks', rc=custom_params)
```

```python
plt.rcParams.keys()
```

```python
FILENAME = "qtip4pf_scan.txt"
qtip4pf = pd.read_csv(f"Data/Force_field_scans/{FILENAME}", delim_whitespace=True)
qtip4pf['Energy'] = qtip4pf['Energy'] - 0.00035782756
```

```python
FILENAME = "schwenke_scan.txt"
schwenke = pd.read_csv(f"Data/Force_field_scans/{FILENAME}", delim_whitespace=True)
```

```python
FILENAME = "cvrqd_scan.txt"
cvrqd = pd.read_csv(f"Data/Force_field_scans/{FILENAME}", delim_whitespace=True)
cvrqd['Energy'] = cvrqd['Energy'] - 0.0035885049
```

```python
# Extract columns
# Convert bond lengths to Angstrom
# Convert energies to eV
qtip4pf_bond_length = qtip4pf["rOH"]*0.529177
qtip4pf_energy = qtip4pf["Energy"]*27.2114
schwenke_bond_length = schwenke["rOH"]*0.529177
schwenke_energy = schwenke["Energy"]*27.2114
cvrqd_bond_length = cvrqd["rOH"]*0.529177
cvrqd_energy = cvrqd["Energy"]*27.2114
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
sns.lineplot(x=qtip4pf_bond_length, y=qtip4pf_energy, ax=ax, label="qTIP4P/F", color="#000004")
sns.lineplot(x=schwenke_bond_length, y=schwenke_energy, ax=ax, label="Schwenke", linestyle="--", color="#51127c")
sns.lineplot(x=cvrqd_bond_length, y=cvrqd_energy, ax=ax, label="CVRQD", linestyle=":", color="#fc8961")

plt.xlabel(r'O-H Bond Length / $\AA$')
plt.xlim(0.7, 2.1)
plt.ylabel("Energy / eV")
plt.ylim(0, 6)
plt.legend()

plt.tight_layout()
plt.savefig("Plots/force_field_scans_full.png", dpi=300, bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
sns.lineplot(x=qtip4pf_bond_length, y=qtip4pf_energy, ax=ax, label="qTIP4P/F", color="#000004")
sns.lineplot(x=schwenke_bond_length, y=schwenke_energy, ax=ax, label="Schwenke", linestyle="--", color="#51127c")
sns.lineplot(x=cvrqd_bond_length, y=cvrqd_energy, ax=ax, label="CVRQD", linestyle=":", color="#fc8961")

plt.xlabel(r'O-H Bond Length / $\AA$')
plt.xlim(0.8, 1.15)
plt.ylabel("Energy / eV")
plt.ylim(0, 0.5)
plt.legend()

plt.tight_layout()
plt.savefig("Plots/force_field_scans_minima.png", dpi=300, bbox_inches="tight")
```
